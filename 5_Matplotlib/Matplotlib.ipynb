```python
#Matplotlib is the most popular plotting library for pythin 
#It gives you control over every aspect of a figure
#It was designed to have a similar feel to Matlab's graphical plotting
#Goto MatplotLib website then gallery to find the relevant graphs
import matplotlib.pyplot as plt
%matplotlib inline #for showing plots in the jupyter notebook
#plt.show()
```

```python
import numpy as np
x = np.linspace(0,5,11)
y=x ** 2
```

```python
x
```

```python
y
```

```python
#Funtional Method of creating plots
plt.plot(x, y, 'r-')
plt.xlabel('X Label')
plt.ylabel('Y label')
plt.title('Title')
```

```python
#Multiplots in the same canvas
plt.subplot(1,2,1) #1 row by two column and plot number=1
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#Object oriented method
fig=plt.figure()
```

```python
axes=fig.add_axes([0.1, 0.1, 0.8, 0.8]) #percent of the blank canvas
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
%matplotlib inline
```

```python
fig=plt.figure()
axis1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #left, bottom, width and height from the orginal canvas
axis2= fig.add_axes([0.2, 0.5, 0.4, 0.3]) #respective to the previous canvas


axis1.plot(x,y)
axis1.set_title("LARGER PLOT")
axis2.plot(y,x)
axis2.set_title("SMALLER PLOT")
```

```python
fig,axes=plt.subplots(nrows=1, ncols=2) #fig represents the canvas and axes represents the canvas objects 
plt.tight_layout() #fixes overlapping
```

```python
fig
```

```python
axes
```

```python
fig,axis=plt.subplots(nrows=1, ncols=2)
for current_ax in axis:
    current_ax.plot(x,y)
    %matplotlib inline
```

```python
fig,axis=plt.subplots(nrows=1, ncols=2)
axis[0].plot(x,y)
axis[0].set_title("Firset Plot")
axis[1].plot(y,x)
axis[1].set_title("Second Plot")

plt.tight_layout()
```

```python
fig=plt.figure(figsize=(3,2)) #three by two inch

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, ax=plt.subplots(nrows=2, ncols=1, figsize=(3,2)) #three by two inch
ax[0].plot(x,y)
ax[1].plot(y,x)
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label="X Squared")
ax.plot(x,x**3, label="X Cubed")

ax.legend(loc=0)
#ax.legend(loc=[0.1, 0.1])
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.plot(x,y,linewidth=3, color='blue', alpha=0.5, linestyle='-.', marker='o', markersize='20', 
        markerfacecolor='yellow', markeredgewidth=3, markeredgecolor='green') #or rgb hex codes
#linestyle = ls
#linewidth = lw

#ax.set_xline([0,1]) zooms x axis
#ax.set_ylin([0,2]) zooms y axis
```
